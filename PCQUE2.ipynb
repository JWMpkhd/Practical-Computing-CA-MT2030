```python
#Defining a fuction to find the area of the biggestest faces of the cuboid, which calculates the area of all possible faces and compares them to give the final value
def SAOBFOC( l , b , h ):
    A1= l * b

    A2= l * h

    A3= b * h

    if A1 > A2 :
        print("The maximum Face Area of a Cuboid with length",l,"breadth",b,"and height",h,"is :" ,A1)
    elif A2> A3:
        print("The maximum Face Area of a Cuboid with length",l,"breadth",b,"and height",h,"is :" ,A2)
    else :
        print( "The maximum Face Area of a Cuboid with length",l,"breadth",b,"and height",h,"is :", A3)
    return    
```

```python
#The input dimensions of the cuboid
l= [30,40,50]
b=20
h=15
```

```python
#Area of the biggest face of the cuboid
for i in range(len(l)):
    SAOBFOC(l[i],b,h)
    
```
